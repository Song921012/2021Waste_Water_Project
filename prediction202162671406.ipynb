```python
from wbepi import models as md
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from lmfit import Parameters, minimize, report_fit
import pandas as pd
import matplotlib as mp
import scipy.stats as st
import scipy.stats as st

# City information import
City_summary = pd.read_csv("./Data/City_summary.csv")

# City considered
city_index = [13, 14, 17, 25]
for n_view in city_index:
    ## Case and wastewater data import
    city_dealt_with, County_dealt_with, State_dealt_with, Population_dealt_with, gdp = City_summary.loc[n_view,
                                                                                       :].values
    merge_data = pd.read_csv("./Data/" + city_dealt_with + "_merge_data.csv")
    ## predict time range
    predict_index = merge_data["newlyconfirmed"].idxmax()
    ## predict 10 days
    predict_end=predict_index +10
    Fit_data = merge_data.loc[:predict_index, :]
    Train_data = merge_data.loc[predict_index:predict_end, :]
    print("-----------------------" + city_dealt_with + "-------------------")
    # Training Stage
    ## initial values
    initvalue = {"initS": int(Population_dealt_with), "initE": 5, "initA": 5, "initI": 5,
                 "initD": merge_data.loc[0,"confirmed"]-1,
                 "initR1": 0, "initR2": 0,
                 "initW": 2}
    timepara = {"t0": 0, "dt": 1, "tend": int(len(Fit_data.index))}

    ## Parameters

    para_set = Parameters()
    para_set.add("ca", value=0.3, min=0, max=1)
    para_set.add("cr", value=0.3, min=0, max=1)
    para_set.add('N', value=int(Population_dealt_with), vary=False)
    para_set.add('beta', value=0.01, min=0, max=10)  # Estimated
    para_set.add('k_E', value=0.55, vary=False)
    para_set.add('k_A', value=0.55, vary=False)
    para_set.add('sigma', value=1 / 5.8, vary=False)
    para_set.add('rho', value=0.31, vary=False)
    para_set.add('gamma_I', value=1 / 12, vary=False)
    para_set.add('gamma_A', value=1 / 10, vary=False)
    para_set.add('gamma_R', value=1 / 13, vary=False)
    para_set.add('detecting_A', value=0.3, min=0.2, max=0.5)  # estimated
    para_set.add('detecting_I', value=0.5, min=0.5, max=1)  # estimated
    para_set.add('log_p_0', value=8.5, vary=False)  # estimated
    para_set.add('p_E', value=0.1, vary=False)
    para_set.add('p_A', value=0.1, vary=False)
    para_set.add('p_R', value=0.1, vary=False)
    para_set.add('water_volume', value=3000, vary=False)
    para_set.add('waning_rate', value=1.44, vary=False)

    # define error function
    def error(para):
        ctrl1 = lambda t: np.power(para["cr"], t) + para["ca"]
        para_model = md.SEIARW(ctrl=ctrl1, initS=initvalue["initS"], initE=initvalue["initE"], initA=initvalue["initA"],
                               initI=initvalue["initI"], initD=initvalue["initD"], initR1=initvalue["initR1"],
                               initR2=initvalue["initR2"], initW=initvalue["initW"],
                               t0=timepara["t0"], dt=timepara["dt"], tend=timepara["tend"],
                               N=para["N"],
                               beta=para["beta"],
                               k_E=para["k_E"], k_A=para["k_A"], sigma=para["sigma"], rho=para["rho"],
                               gamma_I=para["gamma_I"], gamma_A=para["gamma_A"], gamma_R=para["gamma_R"],
                               detecting_A=para["detecting_A"], detacting_I=para["detecting_I"],
                               p_0=np.power(10, para["log_p_0"]), p_E=para["p_E"], p_A=para["p_A"], p_R=para["p_R"],
                               water_volume=para["water_volume"], waning_rate=para["waning_rate"])
        model_data = para_model.ode_sol()
        A = Fit_data["confirmed"].values > 10
        B = Fit_data["newlyconfirmed"].values > 1
        C = Fit_data["water_virus"].values > 100
        mse_1 = (model_data["solution"][A, 4] - Fit_data["confirmed"].values[A])
        mse_2 = (model_data["newlyconfirmed"][B] - Fit_data["newlyconfirmed"].values[B])
        return np.append(mse_1, 5*mse_2)


    # Parameter estimation
    out = minimize(error, para_set)
    report_fit(out.params)
    print(error(out.params))
    # Show fitting results

    # Parameters

    para_set_p0 = Parameters()
    para_set_p0.add('log_p_0', value=8.5, min=3, max=12)  # estimated


    # estimated

    # define error function
    def error_p0(para):
        ctrl1 = lambda t: np.power(out.params["cr"], t) + out.params["ca"]
        para_model = md.SEIARW(ctrl=ctrl1, initS=initvalue["initS"], initE=initvalue["initE"], initA=initvalue["initA"],
                               initI=initvalue["initI"], initD=initvalue["initD"], initR1=initvalue["initR1"],
                               initR2=initvalue["initR2"], initW=initvalue["initW"],
                               t0=timepara["t0"], dt=timepara["dt"], tend=timepara["tend"],
                               N=out.params["N"],
                               beta=out.params["beta"],
                               k_E=out.params["k_E"], k_A=out.params["k_A"], sigma=out.params["sigma"],
                               rho=out.params["rho"],
                               gamma_I=out.params["gamma_I"], gamma_A=out.params["gamma_A"],
                               gamma_R=out.params["gamma_R"],
                               detecting_A=out.params["detecting_A"], detacting_I=out.params["detecting_I"],
                               p_0=np.exp(para["log_p_0"]), p_E=out.params["p_E"], p_A=out.params["p_A"],
                               p_R=out.params["p_R"],
                               water_volume=out.params["water_volume"], waning_rate=out.params["waning_rate"])
        model_data = para_model.ode_sol()
        mse_3 = model_data["solution"][:, 7] - Fit_data["sm_virus"].values
        return mse_3


    # Parameter estimation
    out_p0 = minimize(error_p0, para_set_p0)
    report_fit(out_p0.params)
    # Show fitting results
    ctrl = lambda t: np.power(out.params["cr"], t) + out.params["ca"]
    result_model = md.SEIARW(ctrl=ctrl, initS=initvalue["initS"], initE=initvalue["initE"], initA=initvalue["initA"],
                             initI=initvalue["initI"], initD=initvalue["initD"], initR1=initvalue["initR1"],
                             initR2=initvalue["initR2"], initW=initvalue["initW"],
                             t0=timepara["t0"], dt=timepara["dt"], tend=timepara["tend"],
                             N=out.params["N"],
                             beta=out.params["beta"],
                             k_E=out.params["k_E"], k_A=out.params["k_A"], sigma=out.params["sigma"],
                             rho=out.params["rho"],
                             gamma_I=out.params["gamma_I"], gamma_A=out.params["gamma_A"],
                             gamma_R=out.params["gamma_R"],
                             detecting_A=out.params["detecting_A"], detacting_I=out.params["detecting_I"],
                             p_0=np.exp(out_p0.params["log_p_0"]), p_E=out.params["p_E"], p_A=out.params["p_A"],
                             p_R=out.params["p_R"],
                             water_volume=out.params["water_volume"], waning_rate=out.params["waning_rate"])
    result_data = result_model.ode_sol()
    print("basic reproduction number:", result_model.BRN())

    plt.figure("Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Train Confirmed-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Fit_data["date"],
             Fit_data.loc[:, 'confirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Fit_data["date"],
             result_data["solution"][:, 4], color="#FFBCFF", lw=4,
             label='Fit')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,result_data["solution"][:, 4],
                                                 result_data["solution"][:, 4]/5)
    plt.fill_between(Fit_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    ax.xaxis.set_major_locator(mp.ticker.MultipleLocator(max(1, int(len(Fit_data.index) / 10))))
    #ax.yaxis.set_major_locator(mp.ticker.MultipleLocator(10)) # 
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with+' Confirmed Cases')
    plt.savefig('./Fig/Prediction/Train/' + city_dealt_with + '_train_Confirmed_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()

    plt.figure("Daily Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Train newly Confirmed-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Fit_data["date"],
             Fit_data.loc[:, 'newlyconfirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Fit_data["date"],
             result_data["newlyconfirmed"], color="#FFBCFF", lw=4,
             label='Fit')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,result_data["newlyconfirmed"],
                                                 result_data["newlyconfirmed"]/5)
    plt.fill_between(Fit_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    ax.xaxis.set_major_locator(mp.ticker.MultipleLocator(max(1, int(len(Fit_data.index) / 10))))
    #ax.yaxis.set_major_locator(mp.ticker.MultipleLocator(10)) # 
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with+' Daily Confirmed Cases')
    plt.savefig('./Fig/Prediction/Train/' + city_dealt_with + '_train_daily_confirmed_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()

    plt.figure("Wasterwater data")
    sns.set_theme(style="darkgrid")
    print("--------------Train Wastewater virus concenntration-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Fit_data["date"],
             Fit_data.loc[:, 'water_virus'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Fit_data["date"],
             result_data["solution"][:, 7], color="#FFBCFF", lw=4,
             label='Fit')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,result_data["solution"][:, 7],
                                                 result_data["solution"][:, 7]/5)
    plt.fill_between(Fit_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    ax.xaxis.set_major_locator(mp.ticker.MultipleLocator(max(1, int(len(Fit_data.index) / 10))))
    #ax.yaxis.set_major_locator(mp.ticker.MultipleLocator(10)) # 
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with+' Wastewater virus concentration')
    plt.savefig('./Fig/Prediction/Train/' + city_dealt_with + '_train_wastewater_virus_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()

    # Prediction

    # initial values
    initvalue_predict = {"initS": result_data["solution"][-1, 0],
                         "initE": result_data["solution"][-1, 1],
                         "initA": result_data["solution"][-1, 2],
                         "initI": result_data["solution"][-1, 3],
                         "initD": result_data["solution"][-1, 4],
                         "initR1": result_data["solution"][-1, 5],
                         "initR2": result_data["solution"][-1, 6],
                         "initW": result_data["solution"][-1, 7]}
    timepara_predict = {"t0": 0, "dt": 1, "tend": int(len(Train_data.index))}

    para_set_predict = Parameters()
    para_set_predict.add("ca", value=out.params["ca"], min=0, max=1)
    para_set_predict.add("cr", value=out.params["cr"], min=0, max=1)
    para_set_predict.add('N', value=int(Population_dealt_with), vary=False)
    para_set_predict.add('beta', value=out.params["beta"], vary=False)  # Estimated
    para_set_predict.add('k_E', value=0.55, vary=False)
    para_set_predict.add('k_A', value=0.55, vary=False)
    para_set_predict.add('sigma', value=1 / 5.8, vary=False)
    para_set_predict.add('rho', value=0.31, vary=False)
    para_set_predict.add('gamma_I', value=1 / 12, vary=False)
    para_set_predict.add('gamma_A', value=1 / 10, vary=False)
    para_set_predict.add('gamma_R', value=1 / 13, vary=False)
    para_set_predict.add('detecting_A', value=out.params["detecting_A"], min=0.2, max=0.5)  # estimated
    para_set_predict.add('detecting_I', value=out.params["detecting_I"], min=0.5, max=1)  # estimated
    para_set_predict.add('log_p_0', value=out_p0.params["log_p_0"], vary=False)  # estimated
    para_set_predict.add('p_E', value=0.1, vary=False)
    para_set_predict.add('p_A', value=0.1, vary=False)
    para_set_predict.add('p_R', value=0.1, vary=False)
    para_set_predict.add('water_volume', value=3000, vary=False)
    para_set_predict.add('waning_rate', value=1.44, vary=False)


    # estimated

    # define error function
    def error_predict(para):
        ctrl1 = lambda t: np.power(para["cr"], t) + para["ca"]
        para_model = md.SEIARW(ctrl=ctrl1, initS=initvalue_predict["initS"],
                               initE=initvalue_predict["initE"],
                               initA=initvalue_predict["initA"],
                               initI=initvalue_predict["initI"],
                               initD=initvalue_predict["initD"],
                               initR1=initvalue_predict["initR1"],
                               initR2=initvalue_predict["initR2"],
                               initW=initvalue_predict["initW"],
                               t0=timepara_predict["t0"], dt=timepara_predict["dt"],
                               tend=timepara_predict["tend"],
                               N=para["N"],
                               beta=para["beta"],
                               k_E=para["k_E"], k_A=para["k_A"], sigma=para["sigma"], rho=para["rho"],
                               gamma_I=para["gamma_I"], gamma_A=para["gamma_A"], gamma_R=para["gamma_R"],
                               detecting_A=para["detecting_A"], detacting_I=para["detecting_I"],
                               p_0=np.power(10, para["log_p_0"]), p_E=para["p_E"], p_A=para["p_A"], p_R=para["p_R"],
                               water_volume=para["water_volume"], waning_rate=para["waning_rate"])
        model_data = para_model.ode_sol()
        #C = Train_data["water_virus"].values > 1
        mse_3 = (model_data["solution"][:, 7] - Train_data["water_virus"].values)
        return mse_3


    # Parameter estimation
    out_predict = minimize(error_predict, para_set_predict)
    report_fit(out_predict.params)
    print(error(out_predict.params))
    # Show fitting results
    ctrl = lambda t: np.power(out_predict.params["cr"], t) + out_predict.params["ca"]
    predict_model = md.SEIARW(ctrl=ctrl, initS=initvalue_predict["initS"],
                              initE=initvalue_predict["initE"],
                              initA=initvalue_predict["initA"],
                              initI=initvalue_predict["initI"],
                              initD=Train_data.loc[predict_index,"confirmed"],
                              initR1=initvalue_predict["initR1"],
                              initR2=initvalue_predict["initR2"],
                              initW=initvalue_predict["initW"],
                              t0=timepara_predict["t0"],
                              dt=timepara_predict["dt"],
                              tend=timepara_predict["tend"],
                              N=out_predict.params["N"],
                              beta=out_predict.params["beta"],
                              k_E=out_predict.params["k_E"],
                              k_A=out_predict.params["k_A"],
                              sigma=out_predict.params["sigma"],
                              rho=out_predict.params["rho"],
                              gamma_I=out_predict.params["gamma_I"],
                              gamma_A=out_predict.params["gamma_A"],
                              gamma_R=out_predict.params["gamma_R"],
                              detecting_A=out_predict.params["detecting_A"],
                              detacting_I=out_predict.params["detecting_I"],
                              p_0=np.exp(out_predict.params["log_p_0"]),
                              p_E=out_predict.params["p_E"],
                              p_A=out_predict.params["p_A"],
                              p_R=out_predict.params["p_R"],
                              water_volume=out_predict.params["water_volume"],
                              waning_rate=out_predict.params["waning_rate"])
    predict_data = predict_model.ode_sol()
    predict_data["newlyconfirmed"][0]= result_data["newlyconfirmed"][-1]
    print("basic reproduction number:", predict_model.BRN())

    plt.figure("Predict Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Predict Confirmed-------------")
    plt.plot(Train_data["date"],
             Train_data.loc[:, 'confirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    print(Train_data["date"])
    plt.plot(Train_data["date"],
             predict_data["solution"][:, 4], color="#E64B35", lw=4,
             label='Prediction')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,predict_data["solution"][:, 4],predict_data["solution"][:, 4]/5)
    plt.fill_between(Train_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with + ' Confirmed Cases')
    plt.savefig('./Fig/Prediction/Predict/' + city_dealt_with + '_predict_Confirmed_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()

    plt.figure("Predict Daily Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Predict Daily Confirmed-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Train_data["date"],
             Train_data.loc[:, 'newlyconfirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Train_data["date"],
             predict_data["newlyconfirmed"], color="#E64B35", lw=4,
             label='Prediction')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,predict_data["newlyconfirmed"],predict_data["newlyconfirmed"]/5)
    plt.fill_between(Train_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')

    plt.xticks(rotation=60)
    ax = plt.gca()
    plt.legend(loc='upper right', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.title(city_dealt_with + ' Daily Confirmed Cases')
    plt.savefig('./Fig/Prediction/Predict/' + city_dealt_with + '_predict_daily_confirmed_fit202162291315.png', dpi=100,
                bbox_inches='tight')
    plt.show()
```

```python
from wbepi import models as md
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from lmfit import Parameters, minimize, report_fit
import pandas as pd
import matplotlib as mp
import scipy.stats as st

# City information import
City_summary = pd.read_csv("./Data/City_summary.csv")

# City considered
city_index = [0] # Laramie
for n_view in city_index:
    ## Case and wastewater data import
    city_dealt_with, County_dealt_with, State_dealt_with, Population_dealt_with, gdp = City_summary.loc[n_view,
                                                                                       :].values
    merge_data = pd.read_csv("./Data/" + city_dealt_with + "_merge_data.csv")
    merge_data = merge_data[merge_data["water_volume"]>40000]
    merge_data.index=range(len(merge_data.index))
    merge_data=merge_data.drop([41])
    merge_data.index=range(len(merge_data.index))
    ## predict time range
    predict_index = merge_data["newlyconfirmed"].idxmax()
    ## predict 10 days
    predict_end=predict_index +10
    Fit_data = merge_data.loc[:predict_index, :]
    Train_data = merge_data.loc[predict_index:predict_end, :]
    print("-----------------------" + city_dealt_with + "-------------------")
    # Training Stage
    ## initial values
    initvalue = {"initS": int(Population_dealt_with), "initE": 5, "initA": 5, "initI": 5,
                 "initD": merge_data.loc[0,"confirmed"]-1,
                 "initR1": 0, "initR2": 0,
                 "initW": 2}
    timepara = {"t0": 0, "dt": 1, "tend": int(len(Fit_data.index))}

    ## Parameters

    para_set = Parameters()
    para_set.add("ca", value=0.3, min=0, max=1)
    para_set.add("cr", value=0.3, min=0, max=1)
    para_set.add('N', value=int(Population_dealt_with), vary=False)
    para_set.add('beta', value=0.01, min=0, max=10)  # Estimated
    para_set.add('k_E', value=0.55, vary=False)
    para_set.add('k_A', value=0.55, vary=False)
    para_set.add('sigma', value=1 / 5.8, vary=False)
    para_set.add('rho', value=0.31, vary=False)
    para_set.add('gamma_I', value=1 / 12, vary=False)
    para_set.add('gamma_A', value=1 / 10, vary=False)
    para_set.add('gamma_R', value=1 / 13, vary=False)
    para_set.add('detecting_A', value=0.3, min=0.2, max=0.5)  # estimated
    para_set.add('detecting_I', value=0.5, min=0.5, max=1)  # estimated
    para_set.add('log_p_0', value=8.5, vary=False)  # estimated
    para_set.add('p_E', value=0.1, vary=False)
    para_set.add('p_A', value=0.1, vary=False)
    para_set.add('p_R', value=0.1, vary=False)
    para_set.add('water_volume', value=3000, vary=False)
    para_set.add('waning_rate', value=1.44, vary=False)

    # define error function
    def error(para):
        ctrl1 = lambda t: np.power(para["cr"], t) + para["ca"]
        para_model = md.SEIARW(ctrl=ctrl1, initS=initvalue["initS"], initE=initvalue["initE"], initA=initvalue["initA"],
                               initI=initvalue["initI"], initD=initvalue["initD"], initR1=initvalue["initR1"],
                               initR2=initvalue["initR2"], initW=initvalue["initW"],
                               t0=timepara["t0"], dt=timepara["dt"], tend=timepara["tend"],
                               N=para["N"],
                               beta=para["beta"],
                               k_E=para["k_E"], k_A=para["k_A"], sigma=para["sigma"], rho=para["rho"],
                               gamma_I=para["gamma_I"], gamma_A=para["gamma_A"], gamma_R=para["gamma_R"],
                               detecting_A=para["detecting_A"], detacting_I=para["detecting_I"],
                               p_0=np.power(10, para["log_p_0"]), p_E=para["p_E"], p_A=para["p_A"], p_R=para["p_R"],
                               water_volume=para["water_volume"], waning_rate=para["waning_rate"])
        model_data = para_model.ode_sol()
        A = Fit_data["confirmed"].values > 10
        B = Fit_data["newlyconfirmed"].values > 1
        C = Fit_data["water_virus"].values > 100
        mse_1 = (model_data["solution"][A, 4] - Fit_data["confirmed"].values[A])
        mse_2 = (model_data["newlyconfirmed"][B] - Fit_data["newlyconfirmed"].values[B])
        return np.append(mse_1, 5*mse_2)


    # Parameter estimation
    out = minimize(error, para_set)
    report_fit(out.params)
    print(error(out.params))
    # Show fitting results

    # Parameters

    para_set_p0 = Parameters()
    para_set_p0.add('log_p_0', value=8.5, min=3, max=12)  # estimated


    # estimated

    # define error function
    def error_p0(para):
        ctrl1 = lambda t: np.power(out.params["cr"], t) + out.params["ca"]
        para_model = md.SEIARW(ctrl=ctrl1, initS=initvalue["initS"], initE=initvalue["initE"], initA=initvalue["initA"],
                               initI=initvalue["initI"], initD=initvalue["initD"], initR1=initvalue["initR1"],
                               initR2=initvalue["initR2"], initW=initvalue["initW"],
                               t0=timepara["t0"], dt=timepara["dt"], tend=timepara["tend"],
                               N=out.params["N"],
                               beta=out.params["beta"],
                               k_E=out.params["k_E"], k_A=out.params["k_A"], sigma=out.params["sigma"],
                               rho=out.params["rho"],
                               gamma_I=out.params["gamma_I"], gamma_A=out.params["gamma_A"],
                               gamma_R=out.params["gamma_R"],
                               detecting_A=out.params["detecting_A"], detacting_I=out.params["detecting_I"],
                               p_0=np.exp(para["log_p_0"]), p_E=out.params["p_E"], p_A=out.params["p_A"],
                               p_R=out.params["p_R"],
                               water_volume=out.params["water_volume"], waning_rate=out.params["waning_rate"])
        model_data = para_model.ode_sol()
        mse_3 = model_data["solution"][:, 7] - Fit_data["sm_virus"].values
        return mse_3


    # Parameter estimation
    out_p0 = minimize(error_p0, para_set_p0)
    report_fit(out_p0.params)
    # Show fitting results
    ctrl = lambda t: np.power(out.params["cr"], t) + out.params["ca"]
    result_model = md.SEIARW(ctrl=ctrl, initS=initvalue["initS"], initE=initvalue["initE"], initA=initvalue["initA"],
                             initI=initvalue["initI"], initD=initvalue["initD"], initR1=initvalue["initR1"],
                             initR2=initvalue["initR2"], initW=initvalue["initW"],
                             t0=timepara["t0"], dt=timepara["dt"], tend=timepara["tend"],
                             N=out.params["N"],
                             beta=out.params["beta"],
                             k_E=out.params["k_E"], k_A=out.params["k_A"], sigma=out.params["sigma"],
                             rho=out.params["rho"],
                             gamma_I=out.params["gamma_I"], gamma_A=out.params["gamma_A"],
                             gamma_R=out.params["gamma_R"],
                             detecting_A=out.params["detecting_A"], detacting_I=out.params["detecting_I"],
                             p_0=np.exp(out_p0.params["log_p_0"]), p_E=out.params["p_E"], p_A=out.params["p_A"],
                             p_R=out.params["p_R"],
                             water_volume=out.params["water_volume"], waning_rate=out.params["waning_rate"])
    result_data = result_model.ode_sol()
    print("basic reproduction number:", result_model.BRN())

    plt.figure("Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Train Confirmed-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Fit_data["date"],
             Fit_data.loc[:, 'confirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Fit_data["date"],
             result_data["solution"][:, 4], color="#FFBCFF", lw=4,
             label='Fit')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,result_data["solution"][:, 4],
                                                 result_data["solution"][:, 4]/5)
    plt.fill_between(Fit_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    ax.xaxis.set_major_locator(mp.ticker.MultipleLocator(max(1, int(len(Fit_data.index) / 10))))
    #ax.yaxis.set_major_locator(mp.ticker.MultipleLocator(10)) # 
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with+' Confirmed Cases')
    plt.savefig('./Fig/Prediction/Train/' + city_dealt_with + '_train_Confirmed_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()

    plt.figure("Daily Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Train newly Confirmed-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Fit_data["date"],
             Fit_data.loc[:, 'newlyconfirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Fit_data["date"],
             result_data["newlyconfirmed"], color="#FFBCFF", lw=4,
             label='Fit')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,result_data["newlyconfirmed"],
                                                 result_data["newlyconfirmed"]/5)
    plt.fill_between(Fit_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    ax.xaxis.set_major_locator(mp.ticker.MultipleLocator(max(1, int(len(Fit_data.index) / 10))))
    #ax.yaxis.set_major_locator(mp.ticker.MultipleLocator(10)) # 
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with+' Daily Confirmed Cases')
    plt.savefig('./Fig/Prediction/Train/' + city_dealt_with + '_train_daily_confirmed_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()

    plt.figure("Wasterwater data")
    sns.set_theme(style="darkgrid")
    print("--------------Train Wastewater virus concenntration-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Fit_data["date"],
             Fit_data.loc[:, 'water_virus'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Fit_data["date"],
             result_data["solution"][:, 7], color="#FFBCFF", lw=4,
             label='Fit')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,result_data["solution"][:, 7],
                                                 result_data["solution"][:, 7]/5)
    plt.fill_between(Fit_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    ax.xaxis.set_major_locator(mp.ticker.MultipleLocator(max(1, int(len(Fit_data.index) / 10))))
    #ax.yaxis.set_major_locator(mp.ticker.MultipleLocator(10)) # 
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with+' Wastewater virus concentration')
    plt.savefig('./Fig/Prediction/Train/' + city_dealt_with + '_train_wastewater_virus_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()
    # Prediction

    # initial values
    initvalue_predict = {"initS": result_data["solution"][-1, 0],
                         "initE": result_data["solution"][-1, 1],
                         "initA": result_data["solution"][-1, 2],
                         "initI": result_data["solution"][-1, 3],
                         "initD": result_data["solution"][-1, 4],
                         "initR1": result_data["solution"][-1, 5],
                         "initR2": result_data["solution"][-1, 6],
                         "initW": result_data["solution"][-1, 7]}
    timepara_predict = {"t0": 0, "dt": 1, "tend": int(len(Train_data.index))}

    para_set_predict = Parameters()
    para_set_predict.add("ca", value=out.params["ca"], min=0, max=1)
    para_set_predict.add("cr", value=out.params["cr"], min=0, max=1)
    para_set_predict.add('N', value=int(Population_dealt_with), vary=False)
    para_set_predict.add('beta', value=out.params["beta"], vary=False)  # Estimated
    para_set_predict.add('k_E', value=0.55, vary=False)
    para_set_predict.add('k_A', value=0.55, vary=False)
    para_set_predict.add('sigma', value=1 / 5.8, vary=False)
    para_set_predict.add('rho', value=0.31, vary=False)
    para_set_predict.add('gamma_I', value=1 / 12, vary=False)
    para_set_predict.add('gamma_A', value=1 / 10, vary=False)
    para_set_predict.add('gamma_R', value=1 / 13, vary=False)
    para_set_predict.add('detecting_A', value=out.params["detecting_A"], min=0.2, max=0.5)  # estimated
    para_set_predict.add('detecting_I', value=out.params["detecting_I"], min=0.5, max=1)  # estimated
    para_set_predict.add('log_p_0', value=out_p0.params["log_p_0"], vary=False)  # estimated
    para_set_predict.add('p_E', value=0.1, vary=False)
    para_set_predict.add('p_A', value=0.1, vary=False)
    para_set_predict.add('p_R', value=0.1, vary=False)
    para_set_predict.add('water_volume', value=3000, vary=False)
    para_set_predict.add('waning_rate', value=1.44, vary=False)


    # estimated

    # define error function
    def error_predict(para):
        ctrl1 = lambda t: np.power(para["cr"], t) + para["ca"]
        para_model = md.SEIARW(ctrl=ctrl1, initS=initvalue_predict["initS"],
                               initE=initvalue_predict["initE"],
                               initA=initvalue_predict["initA"],
                               initI=initvalue_predict["initI"],
                               initD=initvalue_predict["initD"],
                               initR1=initvalue_predict["initR1"],
                               initR2=initvalue_predict["initR2"],
                               initW=initvalue_predict["initW"],
                               t0=timepara_predict["t0"], dt=timepara_predict["dt"],
                               tend=timepara_predict["tend"],
                               N=para["N"],
                               beta=para["beta"],
                               k_E=para["k_E"], k_A=para["k_A"], sigma=para["sigma"], rho=para["rho"],
                               gamma_I=para["gamma_I"], gamma_A=para["gamma_A"], gamma_R=para["gamma_R"],
                               detecting_A=para["detecting_A"], detacting_I=para["detecting_I"],
                               p_0=np.power(10, para["log_p_0"]), p_E=para["p_E"], p_A=para["p_A"], p_R=para["p_R"],
                               water_volume=para["water_volume"], waning_rate=para["waning_rate"])
        model_data = para_model.ode_sol()
        #C = Train_data["water_virus"].values > 1
        mse_3 = (model_data["solution"][:, 7] - Train_data["water_virus"].values)
        return mse_3


    # Parameter estimation
    out_predict = minimize(error_predict, para_set_predict)
    report_fit(out_predict.params)
    print(error(out_predict.params))
    # Show fitting results
    ctrl = lambda t: np.power(out_predict.params["cr"], t) + out_predict.params["ca"]
    predict_model = md.SEIARW(ctrl=ctrl, initS=initvalue_predict["initS"],
                              initE=initvalue_predict["initE"],
                              initA=initvalue_predict["initA"],
                              initI=initvalue_predict["initI"],
                              initD=Train_data.loc[predict_index,"confirmed"],
                              initR1=initvalue_predict["initR1"],
                              initR2=initvalue_predict["initR2"],
                              initW=initvalue_predict["initW"],
                              t0=timepara_predict["t0"],
                              dt=timepara_predict["dt"],
                              tend=timepara_predict["tend"],
                              N=out_predict.params["N"],
                              beta=out_predict.params["beta"],
                              k_E=out_predict.params["k_E"],
                              k_A=out_predict.params["k_A"],
                              sigma=out_predict.params["sigma"],
                              rho=out_predict.params["rho"],
                              gamma_I=out_predict.params["gamma_I"],
                              gamma_A=out_predict.params["gamma_A"],
                              gamma_R=out_predict.params["gamma_R"],
                              detecting_A=out_predict.params["detecting_A"],
                              detacting_I=out_predict.params["detecting_I"],
                              p_0=np.exp(out_predict.params["log_p_0"]),
                              p_E=out_predict.params["p_E"],
                              p_A=out_predict.params["p_A"],
                              p_R=out_predict.params["p_R"],
                              water_volume=out_predict.params["water_volume"],
                              waning_rate=out_predict.params["waning_rate"])
    predict_data = predict_model.ode_sol()
    predict_data["newlyconfirmed"][0]= result_data["newlyconfirmed"][-1]
    print("basic reproduction number:", predict_model.BRN())

    plt.figure("Predict Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Predict Confirmed-------------")
    plt.plot(Train_data["date"],
             Train_data.loc[:, 'confirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    print(Train_data["date"])
    plt.plot(Train_data["date"],
             predict_data["solution"][:, 4], color="#E64B35", lw=4,
             label='Prediction')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,predict_data["solution"][:, 4],predict_data["solution"][:, 4]/10)
    plt.fill_between(Train_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')
    plt.xticks(rotation=60)
    ax = plt.gca()
    plt.legend(loc='upper left', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.title(city_dealt_with + ' Confirmed Cases')
    plt.savefig('./Fig/Prediction/Predict/' + city_dealt_with + '_predict_Confirmed_fit202162291315.png', dpi=100, bbox_inches='tight')
    plt.show()

    plt.figure("Predict Daily Confirmed Case Data")
    sns.set_theme(style="darkgrid")
    print("--------------Predict Daily Confirmed-------------")
    # plt.rcParams['font.sans-serif']=['SimHei'] # chinese support
    plt.plot(Train_data["date"],
             Train_data.loc[:, 'newlyconfirmed'],
             'o', color="#46ACC8", ms=8,
             label='Real')
    plt.plot(Train_data["date"],
             predict_data["newlyconfirmed"], color="#E64B35", lw=4,
             label='Prediction')
    low_CI_bound, high_CI_bound=st.norm.interval(0.95,predict_data["newlyconfirmed"],predict_data["newlyconfirmed"]/5)
    plt.fill_between(Train_data["date"],low_CI_bound, high_CI_bound, high_CI_bound, alpha=0.2,
                label='confidence interval')

    plt.xticks(rotation=60)
    ax = plt.gca()
    plt.legend(loc='upper right', borderaxespad=0.)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.title(city_dealt_with + ' Daily Confirmed Cases')
    plt.savefig('./Fig/Prediction/Predict/' + city_dealt_with + '_predict_daily_confirmed_fit202162291315.png', dpi=100,
                bbox_inches='tight')
    plt.show()
```
